## iloc Summarization

### Single Selections using iloc: -

```python
#Rows: - 
df.iloc[0] #first row of df.
df.iloc[1] #second row of df.
df.iloc[-1] #last row of df.
```

```python
#Columns: - 
df.iloc[:,0] #first coolumn of the df
df.iloc[:,1] #second coolumn of the df
df.iloc[:,-1:] #last coolumn of the df #Colon after -1 will help us to fetch the column label as well.
```

```python
#Multiple row and column selections: - 
#syntax - df.iloc[<row selection ,<column selection>]
df.iloc[0:5] #first five rows (incl. everey column) of the df
df.iloc[:,0:2] #first 2 columns of the df - MAX. used.
df.iloc[0:5,5:8] #first 5 rows (0:5) and 5th, 6th & 7th columns only (5:8) will be sliced from the parent df.
df.iloc[[0,3,6,24],[0,1,2]] # 1st, 4th, 7th, 25th Rows (0,3,6,24) of the df and 1st, 2nd & 3rd columns (0,1,2) of my df.
```
